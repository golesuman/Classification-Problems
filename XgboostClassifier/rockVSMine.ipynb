```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
pd.set_option('display.max_columns',None)
```

```python
df=pd.read_csv('Copy of sonar data.csv',header=None)
df.head()
```

```python
df[60].value_counts()
```

```python
numeric=[feature for feature in df.columns if df[feature].dtypes!='O']
for feature in numeric:
    sns.distplot(df[feature])
    plt.show()
```

```python
sns.heatmap(df.isnull())
```

## Since there is no null value we can actually do feature engineering

```python
x=df.drop(columns=60,axis=1)
y=df[60]
y.head()
```

```python
## Choose model from multiple models
from sklearn.linear_model import LogisticRegression
from xgboost import XGBClassifier
from sklearn.ensemble import RandomForestClassifier
models=[LogisticRegression(),RandomForestClassifier(),XGBClassifier()]
from sklearn.model_selection import cross_val_score,KFold
from sklearn.model_selection import RandomizedSearchCV
```

```python
for model in models:
    print(f'{model}:{np.mean(cross_val_score(model,X=x,y=y,cv=6))}')
    
```

```python
params={
    'n_iter':[np.linspace(1,500,10)],
}
clf=RandomizedSearchCV(XGBClassifier(),params,cv=6,return_train_score=False)
```

```python
xg=clf.fit(x,y)
```

```python
xg.best_estimator_
```

```python
df_results=pd.DataFrame(xg.cv_results_)
df_results
```

```python
y_pred=xg.predict(x)
y_pred
```

```python
y
```

```python
from sklearn.metrics import accuracy_score
acc=accuracy_score(y,y_pred)
```
