```python
df_1=df.drop(['class'],axis=1)
x=df_1.loc[:1500]
df_y=df['class']
y=df_y.loc[:1500]
x.head()
```

```python
params={'C':[10,20,30,40],
    'kernel':['rbf','linear']
    }
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
grid=GridSearchCV(SVC(gamma='auto'),params,cv=5,return_train_score=False)
```

```python
grid.fit(x,y)
```

```python
df_results=pd.DataFrame(grid.cv_results_)
df_results
```

```python
df_score=pd.DataFrame(df_results[['params','mean_test_score','rank_test_score']])
df_score
```

```python
model=SVC(C=40,kernel='rbf',gamma='auto')
model.fit(x,y)
import pickle 
pickle.dump(model,open('car_model.pkl','wb'))
```
